```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
lit = pd.read_csv('../DATA/lit.csv')
san = pd.read_csv('../DATA/san.csv')
```

```python
df = pd.merge(lit,san,on = "State/UT")
df.head()
```

```python
df.to_csv(r"C:\Users\Anish\OneDrive\Desktop\minor\DATA\final_result.csv", index=False)
```

```python
r = df["Total Literacy"].corr(df["Percentage with basic sanitation service"])
print("Correlation (r):", r)
```

```python
r_squared = r ** 2
print(r_squared)
```

```python
plt.scatter(df["Total Literacy"], df["Percentage with basic sanitation service"])
m, b = np.polyfit(df["Total Literacy"], df["Percentage with basic sanitation service"], 1)
plt.plot(df["Total Literacy"], m*df["Total Literacy"] + b)
plt.xlabel("Total Literacy")
plt.ylabel("Basic Sanitation")
plt.title("Literacy vs Sanitation")
plt.text(70, 95, f"r = {r:.2f}\nR² = {r**2:.2f}")
plt.show()
```

There is a moderately strong positive association between total literacy and basic sanitation coverage across Indian states (r = 0.67). Approximately 45% of the variation in sanitation coverage can be associated with literacy levels. However, the presence of outliers suggests that additional structural and geographical factors influence sanitation outcomes.
