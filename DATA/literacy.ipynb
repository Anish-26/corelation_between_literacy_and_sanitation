```python
import pandas as pd
import numpy as numpy
```

```python
df = pd.read_csv("women.csv")
df.head()
```

```python
df = df.rename(columns={'Percentage literate': 'Percentage literate (women)'})
df.head()
```

```python
df1 = pd.read_csv("men.csv")
df1.head()
```

```python
df1 = df1.rename(columns={'Percentage literate': 'Percentage literate (men)'})
df1.head()
```

```python
df3 = pd.merge(df,df1,on = "State/union territory")
df3.head()
```

```python
df3.info()
```

```python
df3 = df3.drop(["No schooling_x","<5 years complete_x","5-7 years complete_x","8-9 years complete_x","10-11 years complete_x","12 or more years complete_x","Total_x","No schooling_y","<5 years complete_y","5-7 years complete_y","8-9 years complete_y","10-11 years complete_y","12 or more years complete_y","Total_y"],axis = 1)
df3 = df3.drop(0)
df3.head()
```

```python
df3['Literacy Gap'] = df3['Percentage literate (men)'] - df3['Percentage literate (women)']
df3["Total Literacy"] = (df3['Percentage literate (men)']+df3['Percentage literate (women)'])/2
df3.head()
```

```python
df3 = df3.drop(columns=["Percentage literate (women)","Percentage literate (men)","Literacy Gap"],axis = 1)
df3.head()
```

```python
df3 = df3.rename(columns = {"State/union territory":"State/UT"})
df3.head()
```
